```python
# WEEK 5: STATISTICAL ANALYSIS

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#  DESCRIPTIVE STATISTICS
print("\nDESCRIPTIVE STATISTICS")

print("Average Salary Change (%):", df["salary_change_percent"].mean())
print("Median Salary Change (%):", df["salary_change_percent"].median())
print("Standard Deviation:", df["salary_change_percent"].std())
print("Minimum:", df["salary_change_percent"].min())
print("Maximum:", df["salary_change_percent"].max())
```

```python
#  HISTOGRAM
plt.figure(figsize=(8,5))
plt.hist(df["salary_change_percent"], bins=30)
plt.title("Salary Change Percentage Distribution")
plt.xlabel("Salary Change (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
#  BOXPLOT (Outlier Detection)
plt.figure(figsize=(6,4))
sns.boxplot(x=df["salary_change_percent"])
plt.title("Salary Change Percentage Boxplot")
plt.show()
```

```python
#  NORMALITY TEST
# ------------------------------
stat, p = stats.shapiro(df["salary_change_percent"])

print("\nShapiro-Wilk Test p-value:", p)

if p > 0.05:
    print("Data is normally distributed")
else:
    print("Data is NOT normally distributed")
```

```python
#  CORRELATION MATRIX
# ------------------------------
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
